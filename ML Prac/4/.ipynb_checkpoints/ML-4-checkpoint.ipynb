```python
x=2 # starting point
lr = 0.01    #learning rate
precision =0.000001 #to stop the algorithm
previous_step_size =1
max_iter=1000
iters=0
gf=lambda x: (x+3)**2 #gradient function

```

```python
import matplotlib.pyplot as plt
```

```python
gd=[]
```

```python
while precision < previous_step_size and iters<max_iter:
    prev = x;
    X = x-lr*gf(prev)
    previous_step_size =abs(X - prev)
    iters +=  1
    print('Iteration :',iters,'Value : ', x)
    gd.append(x)
```

```python
print('Local minima : ' ,X) #last value
```

```python
plt.plot(gd)
```
